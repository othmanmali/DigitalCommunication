# ََQPSK:

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
bits = np.random.randint(2, size=(1, 20))
bits
```

```python
bits = bits.reshape(10,2)
```

```python
bits
```

```python
factor = np.array([[1], [2]])
new_bits = bits.dot(factor)
new_bits
```

```python
new_bits = new_bits.reshape(1,10)
new_bits
```

```python
t = np.arange(0, 20, 0.02)
f = 1
```

```python
carrier = np.sin(2*pi*f*t)
```

```python
plt.figure(figsize=(10, 5), dpi=100)
plt.plot(carrier)
```

```python
phases = (new_bits * pi) / 2
phases
```

```python
rep_phases = np.repeat(phases, 100)
```

```python
bits_mod = np.sin((2*pi*f*t) + rep_phases)
```

```python
x = [i for i in range(2,20,2)]
plt.figure(figsize=(10, 5), dpi=100)
plt.vlines(x, -1.5, 1.5, color='red', linewidth=2, linestyles='dashed')
plt.title('{} \n \n with T = 2 sec'.format(bits.reshape(1,20)))
plt.plot(t, bits_mod)
```
